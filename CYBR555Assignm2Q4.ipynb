```python
import networkx as nx
import matplotlib.pyplot as plt

# Create directed graph
G = nx.DiGraph()

# Let's add nodes
nodes = ["S1", "S2", "S3", "F1", "F2", "P1", "P2", "D1", "D2"]
G.add_nodes_from(nodes)

# Let's add edges (with access rights)
edges = [
    ("S1", "S1", "control"),
    ("S1", "S2", "owner"),
    ("S1", "S3", "owner, control"),
    ("S1", "F1", "read*"),
    ("S1", "F2", "read, owner"),
    ("S1", "P1", "wakeup"),
    ("S1", "P2", "wakeup"),
    ("S1", "D1", "seek"),
    ("S1", "D2", "owner"),
    ("S2", "S1", "control"),
    ("S2", "F1", "write*"),
    ("S2", "F2", "execute"),
    ("S2", "D1", "owner"),
    ("S2", "D2", "seek*"),
    ("S3", "S1", "control"),
    ("S3", "F1", "write"),
    ("S3", "P1", "stop"),
]

# Let's add the edges to the graph
for src, dst, label in edges:
    G.add_edge(src, dst, label=label)

# Let's draw the graph
plt.figure(figsize=(20, 12))
pos = nx.spring_layout(G, seed=42)
edge_labels = {(src, dst): lbl for src, dst, lbl in edges}
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=10)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

# Show graph
plt.title("Discretionary Access Control (DAC) Directed Graph")
plt.show()
```
